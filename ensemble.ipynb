```python
import pandas as pd

data1 = pd.read_csv('task1_E.csv')
data2 = pd.read_csv('task2_E.csv')
data3 = pd.read_csv('task3_E.csv')
data1_2 = pd.read_csv('task1_ST.csv')
data2_2 = pd.read_csv('task2_ST.csv')
data3_2 = pd.read_csv('task3_ST.csv')

ENSEMBLE_CHOICE = int(input("Which ensemble combination to output?\n[1] 5 Classifiers\n[2] 7 Classifiers\n[3] 9 Classifiers"))
```

```python
# Initialize Ensembles
ensemble_1_1 = pd.DataFrame({'bertb': data1['bertb'].tolist(), 
                           'germanbert': data1['germanbert'].tolist(), 
                           'rlgermanbert': data1['rlgermanbert'].tolist(),
                           'bertweet': data1_2['bertweet'].tolist(),
                           'mbertcg': data1['mbertcg'].tolist()})
ensemble_1_2 = pd.DataFrame({'bertb': data2['bertb'].tolist(), 
                           'germanbert': data2['germanbert'].tolist(), 
                           'rlgermanbert': data2['rlgermanbert'].tolist(),
                           'bertweet': data2_2['bertweet'].tolist(),
                           'mbertcg': data2['mbertcg'].tolist()})
ensemble_1_3 = pd.DataFrame({'bertb': data3['bertb'].tolist(), 
                           'germanbert': data3['germanbert'].tolist(), 
                           'rlgermanbert': data3['rlgermanbert'].tolist(),
                           'bertweet': data3_2['bertweet'].tolist(),
                           'mbertcg': data3['mbertcg'].tolist()})
ensemble_2_1 = pd.DataFrame({'bertb': data1['bertb'].tolist(), 
                           'germanbert': data1['germanbert'].tolist(), 
                           'rlgermanbert': data1['rlgermanbert'].tolist(),
                           'bertweet': data1_2['bertweet'].tolist(),
                           'mbertcg': data1['mbertcg'].tolist(),
                           'mbertc': data1['mbertc'].tolist(),
                           'txlmr_de': data1_2['txlmr_de'].tolist()})
ensemble_2_2 = pd.DataFrame({'bertb': data2['bertb'].tolist(), 
                           'germanbert': data2['germanbert'].tolist(), 
                           'rlgermanbert': data2['rlgermanbert'].tolist(),
                           'bertweet': data2_2['bertweet'].tolist(),
                           'mbertcg': data2['mbertcg'].tolist(),
                           'mbertc': data2['mbertc'].tolist(),
                           'txlmr_de': data2_2['txlmr_de'].tolist()})
ensemble_2_3 = pd.DataFrame({'bertb': data3['bertb'].tolist(), 
                           'germanbert': data3['germanbert'].tolist(), 
                           'rlgermanbert': data3['rlgermanbert'].tolist(),
                           'bertweet': data3_2['bertweet'].tolist(),
                           'mbertcg': data3['mbertcg'].tolist(),
                           'mbertc': data3['mbertc'].tolist(),
                           'txlmr_de': data3_2['txlmr_de'].tolist()})
ensemble_3_1 = pd.DataFrame({'bertb': data1['bertb'].tolist(), 
                           'germanbert': data1['germanbert'].tolist(), 
                           'rlgermanbert': data1['rlgermanbert'].tolist(),
                           'bertweet': data1_2['bertweet'].tolist(),
                           'mbertcg': data1['mbertcg'].tolist(),
                           'mbertc': data1['mbertc'].tolist(),
                           'txlmr_de': data1_2['txlmr_de'].tolist(),
                           'txlmr_en': data1_2['txlmr_en'].tolist(),
                           'xlmr_de': data1_2['xlmr_de'].tolist()})
ensemble_3_2 = pd.DataFrame({'bertb': data2['bertb'].tolist(), 
                           'germanbert': data2['germanbert'].tolist(), 
                           'rlgermanbert': data2['rlgermanbert'].tolist(),
                           'bertweet': data2_2['bertweet'].tolist(),
                           'mbertcg': data2['mbertcg'].tolist(),
                           'mbertc': data2['mbertc'].tolist(),
                           'txlmr_de': data2_2['txlmr_de'].tolist(),
                           'txlmr_en': data2_2['txlmr_en'].tolist(),
                           'xlmr_de': data2_2['xlmr_de'].tolist()})
ensemble_3_3 = pd.DataFrame({'bertb': data3['bertb'].tolist(), 
                           'germanbert': data3['germanbert'].tolist(), 
                           'rlgermanbert': data3['rlgermanbert'].tolist(),
                           'bertweet': data3_2['bertweet'].tolist(),
                           'mbertcg': data3['mbertcg'].tolist(),
                           'mbertc': data3['mbertc'].tolist(),
                           'txlmr_de': data3_2['txlmr_de'].tolist(),
                           'txlmr_en': data3_2['txlmr_en'].tolist(),
                           'xlmr_de': data3_2['xlmr_de'].tolist()})

if ENSEMBLE_CHOICE == 1 or ENSEMBLE_CHOICE == 5:
  ## Ensemble 1
  sub1 = []
  sub2 = []
  sub3 = []
  for index, row in ensemble_1_1.iterrows():
    if row['bertb'] + row['germanbert'] + row['rlgermanbert'] + row['bertweet'] + row['mbertcg'] > 2:
      sub1.append(1)
    else:
      sub1.append(0)
  for index, row in ensemble_1_2.iterrows():
    if row['bertb'] + row['germanbert'] + row['rlgermanbert'] + row['bertweet'] + row['mbertcg'] > 2:
      sub2.append(1)
    else:
      sub2.append(0)
  for index, row in ensemble_1_3.iterrows():
    if row['bertb'] + row['germanbert'] + row['rlgermanbert'] + row['bertweet'] + row['mbertcg'] > 2:
      sub3.append(1)
    else:
      sub3.append(0)
  result_1 = pd.DataFrame({'comment_id': data1['comment_id'], 'Sub1_Toxic': sub1, 'Sub2_Engaging': sub2, 'Sub3_FactClaiming': sub3})
  result_1.to_csv('answer.csv', index=False)
elif ENSEMBLE_CHOICE == 2 or ENSEMBLE_CHOICE == 7:
  ## Ensemble 2
  sub1 = []
  sub2 = []
  sub3 = []
  for index, row in ensemble_2_1.iterrows():
    if row['bertb'] + row['germanbert'] + row['rlgermanbert'] + row['bertweet'] + row['mbertcg'] + row['mbertc'] + row['txlmr_de'] > 3:
      sub1.append(1)
    else:
      sub1.append(0)
  for index, row in ensemble_2_2.iterrows():
    if row['bertb'] + row['germanbert'] + row['rlgermanbert'] + row['bertweet'] + row['mbertcg'] + row['mbertc'] + row['txlmr_de'] > 3:
      sub2.append(1)
    else:
      sub2.append(0)
  for index, row in ensemble_2_3.iterrows():
    if row['bertb'] + row['germanbert'] + row['rlgermanbert'] + row['bertweet'] + row['mbertcg'] + row['mbertc'] + row['txlmr_de'] > 3:
      sub3.append(1)
    else:
      sub3.append(0)
  result_2 = pd.DataFrame({'comment_id': data2['comment_id'], 'Sub1_Toxic': sub1, 'Sub2_Engaging': sub2, 'Sub3_FactClaiming': sub3})
  result_2.to_csv('answer.csv', index=False)
elif ENSEMBLE_CHOICE == 3 or ENSEMBLE_CHOICE == 9:
  ## Ensemble 3
  sub1 = []
  sub2 = []
  sub3 = []
  for index, row in ensemble_3_1.iterrows():
    if row['bertb'] + row['germanbert'] + row['rlgermanbert'] + row['bertweet'] + row['mbertcg'] + row['mbertc'] + row['txlmr_de'] + row['txlmr_en'] + row['xlmr_de'] > 4:
      sub1.append(1)
    else:
      sub1.append(0)
  for index, row in ensemble_3_2.iterrows():
    if row['bertb'] + row['germanbert'] + row['rlgermanbert'] + row['bertweet'] + row['mbertcg'] + row['mbertc'] + row['txlmr_de'] + row['txlmr_en'] + row['xlmr_de'] > 4:
      sub2.append(1)
    else:
      sub2.append(0)
  for index, row in ensemble_3_3.iterrows():
    if row['bertb'] + row['germanbert'] + row['rlgermanbert'] + row['bertweet'] + row['mbertcg'] + row['mbertc'] + row['txlmr_de'] + row['txlmr_en'] + row['xlmr_de'] > 4:
      sub3.append(1)
    else:
      sub3.append(0)
  result_3 = pd.DataFrame({'comment_id': data3['comment_id'], 'Sub1_Toxic': sub1, 'Sub2_Engaging': sub2, 'Sub3_FactClaiming': sub3})
  result_3.to_csv('answer.csv', index=False)
```
